```python
# chunk documents
from gitsource import chunk_documents
chunks = chunk_documents(documents, size=2000, step=1000)
```

```python
print(chunks[0])
len(chunks)
```

```python
filtered_chunks = list(filter(lambda chunk: chunk.get('content'), chunks))
print(len(filtered_chunks))

texts = [chunk['content'] for chunk in filtered_chunks]
texts[0]
```

```python
# embed chunks
from tqdm.auto import tqdm
import numpy as np

batch_size = 50
vectors = []

for i in tqdm(range(0, len(texts), batch_size)):
    batch = texts[i:i + batch_size]
    batch_vectors = embed.encode_batch(batch)
    vectors.extend(batch_vectors)

X = np.array(vectors)
```

```python
print(len(X))
print(len(X[0]))
```

```python
scores = X.dot(v1)
```

```python
top5 = np.argsort(-scores)[:5]
```

```python
print(scores[top5[0]])
print(filtered_chunks[top5[0]])
```

```python
# vector search with library
from minsearch import VectorSearch

vIndex = VectorSearch(keyword_fields=['filename'])
vIndex.fit(X, filtered_chunks)
```

```python
q2 = 'What metric do we use to evaluate a search engine?'
v2 = embed.encode(q2)
```

```python
results_from_vector = vIndex.search(
    query_vector=v2,
    num_results=5
)
```

```python
# hybrid search
# Each search produces its own ranked list, so we need a way to combine them into one. 
# We use Reciprocal Rank Fusion (RRF). 
# It ignores the raw scores from each method, which live on different scales and aren't directly comparable. 
# Instead, it looks only at the position of each document in each list.

def rrf(result_lists, k=60, num_results=5):
    scores = {}
    docs = {}

    for results in result_lists:
        for rank, doc in enumerate(results):
            key = (doc["filename"], doc["start"])
            scores[key] = scores.get(key, 0) + 1 / (k + rank)
            docs[key] = doc

    ranked = sorted(scores, key=scores.get, reverse=True)
    return [docs[key] for key in ranked[:num_results]]
```

```python
q4 = "How do I give the model access to tools?"
v4 = embed.encode(q4)

vector_results = vIndex.search(v4, num_results=5)
text_results = text_index.search(q4, num_results=5)

results = rrf([vector_results, text_results])
```
